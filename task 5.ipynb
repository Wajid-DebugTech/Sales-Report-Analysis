```python
df['discounted_price'] = df['discounted_price'].replace('[^0-9.]','', regex = True).astype(float)
df['actual_price'] = df['actual_price'].replace('[^0-9.]','', regex = True).astype(float)
df['discount_percentage'] = df['discount_percentage'].replace('%','', regex = True).astype(float)
df['rating_count'] = df['rating_count'].replace('[^0-9.]','', regex = True).astype(float)
```

```python
discrepancy_rows = df[df['rating'].astype(str) == '|']

print(f"Number of rows with just '|': {len(discrepancy_rows)}")

df['rating'] = df['rating'].replace('|', np.nan)

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['rating'].fillna(df['rating'].mean(), inplace = True)
df['rating_count'].fillna(df['rating_count'].median(), inplace = True)
df.isnull().sum()
```

```python
df2 = pd.read_csv("C:/Users/Murtuja afshar/OneDrive/Pictures/Desktop/job/task 5/Average of rating, Average of rating_count and Average of discount_percentage by category groups.csv")
df2.head()
```

```python
#category segregation
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
features = ["Average of rating", "Average of rating_count", "Average of discount_percentage"]
x = df2[features]

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
inertia = []
K = range(1,11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize = (8,5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df2['Clusters'] = kmeans.fit_predict(x_scaled)
df2.head()
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot scatter points, colored by Clusters
scatter = ax.scatter(
    x_scaled[:, 0],  # Scaled Average of rating
    x_scaled[:, 1],  # Scaled Average of rating_count
    x_scaled[:, 2],  # Scaled Average of discount_percentage
    c=df2['Clusters'],  # Color by Clusters column
    cmap='viridis',  # Color map
    s=100,  # Point size
    alpha=0.6  # Transparency
)

# Add labels for each point (abbreviated category names)
for i, txt in enumerate(df2['category (groups)']):
    ax.text(x_scaled[i, 0], x_scaled[i, 1], x_scaled[i, 2], 
            txt.split('|')[0], fontsize=10)

# Set axis labels
ax.set_xlabel('Scaled Average of Rating')
ax.set_ylabel('Scaled Average of Rating Count')
ax.set_zlabel('Scaled Discount Percentage')
ax.set_title('3D Scatter Plot of Categories by Rating, Rating Count, and Discount')

# Add colorbar to show cluster mapping
plt.colorbar(scatter, label='Cluster')

# Show the plot
plt.show()
```
